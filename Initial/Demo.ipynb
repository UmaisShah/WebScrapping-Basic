```python
import requests
from bs4 import BeautifulSoup
from Utils import link
```

## Fetching simple page in Python

```python
try:
  response = requests.get(link)
  if response.status_code == 200:
    print("Successfully retrieved content!")
  else:
    print(f"Error: Status code {response.status_code}")
except requests.exceptions.RequestException as e:
  print(f"An error occurred: {e}")
```

## Formating the raw page to readable form using prettify()

```python
soup = BeautifulSoup(response.text,"html.parser")
print(soup.prettify())
```

## Find Elements by Class Name
### If you want to retrieve data by a particular class name, identifying the right class is critical. The find() command is used to find elements by a particular class or tag name.

### In Chrome, you can check the class name by right-clicking on the required table of the web page and navigating to “Inspect”.

```python
populationTable= soup.find('table', class_='us-census-pop us-census-pop-right')
print(populationTable)
# print(populationTable if populationTable is not None else soup.find('table'))

```

## Find Elements By ID

### In an HTML page, every element can have the id attribute. You can begin to parse your page by selecting a specific element by its ID.

```python
soup.find(id='Population')
```

### To iterate through each row(tr) and assign each element of tr(td) to a variable and add it to the list.

```python
A=[]
B=[]
C=[]
D=[]

for row in populationTable.findAll("tr"):
    cells =row.findAll("td")
    population=row.findAll("th")
    if len(cells)==3:
        A.append(population[0].find(text=True))
        B.append(cells[0].find(text=True))
        C.append(cells[1].find(text=True))
        D.append(cells[2].find(text=True))
```

```python
import pandas as pd
df=pd.DataFrame(A,columns=['Census'])
df['population']=B
df['+-'] = D
df
```
